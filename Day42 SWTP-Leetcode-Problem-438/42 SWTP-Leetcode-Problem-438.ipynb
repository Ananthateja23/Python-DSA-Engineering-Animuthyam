## Leetcode Problem Solution
```python
class Solution:
    def findAnagrams(self, s: str, p: str) -> List[int]:
        def is_anagram(dicia, dicib):
            if len(dicia) != len(dicib):
                return False
            for key in dicia:
                if key not in dicib:
                    return False
                if dicia[key] != dicib[key]:
                    return False
            return True

        dicia = {}
        dicib = {}
        for ch in p:
            if ch in dicib:
                dicib[ch] += 1
            else:
                dicib[ch] = 1
        l = 0
        n = len(s)
        ans = []
        k = len(p)
        for r in range(n):
            val = s[r]
            if val not in dicia:
                dicia[val] = 1
            else:
                dicia[val] += 1
            
            if (r - l == k):
                lval = s[l]
                dicia[lval] -= 1
                if dicia[lval] == 0:
                    dicia.pop(lval)
                l += 1
            
            if (r-l+1 == k):
                if is_anagram(dicia, dicib):
                    ans.append(l)
        return ans
```
